# Topic Description

This portfolio output presents a simple hotel billing system developed using Python functions. It applies fundamental programming concepts, including:

• User input

• Variables

• Expressions

• Functions

• Parameters and return values

• Conditional statements

The program simulates a guest’s check-in and check-out process by calculating the total hours stayed and the corresponding hotel bill based on the hourly room rate. It also handles overnight stays using simple decision-making logic, demonstrating the practical use of functions in a real-life application.

```python
# Greeting the Guest
def greet_guest(name):
    print("Hello,", name)
    print("Welcome to Hotel 99!")

# Computing Total Hours Stayed
def compute_hours(check_in, check_out):
    
    # Conditional Statement for Overnight Stay
    if check_out <= check_in:
        check_out += 24
# Return Total Hours Stayed
    return check_out - check_in

# Computing Total Bill
def compute_bill(hours_stayed, rate_per_hour):
    return hours_stayed * rate_per_hour

# User Information
name = input("Enter guest name: ")
greet_guest(name)

check_in = float(input("Enter check-in time in decimal hours (example 2.0): "))
check_out = float(input("Enter check-out time in decimal hours (example 7.5): "))
rate_per_hour = float(input("Enter rate per hour: "))

# Computation Process
hours_stayed = compute_hours(check_in, check_out)
bill = compute_bill(hours_stayed, rate_per_hour)

# Output
print("Hours stayed:", hours_stayed)
print("Total bill:", bill)
print("Thank you for staying with Hote 99!")
```

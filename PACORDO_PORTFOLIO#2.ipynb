The program computes the student's average grade and determines the corresponding academic status using simple decision-making logic. It demonstrates the practical application of programming in academic evaluation and automated grade assessment.

```python
# User Input Information
Sub1 = float(input('Subject 1: '))
Sub2 = float(input('Subject 2: '))
Sub3 = float(input('Subject 3: '))
Sub4 = float(input('Subject 4: '))
Sub5 = float(input('Subject 5: '))
Sub6 = float(input('Subject 6: '))
Sub7 = float(input('Subject 7: '))
Sub8 = float(input('Subject 8: '))
Sub9 = float(input('Subject 9: '))

# Computations
total = Sub1 + Sub2 + Sub3 + Sub4 + Sub5 + Sub6 + Sub7 + Sub8 + Sub9
percent_gwa = total / 9

grades = 5.0
if percent_gwa >= 100:
    grades = 1.0
elif percent_gwa >= 98:
    grades = 1.1
elif percent_gwa >= 96:
    grades = 1.25
elif percent_gwa >= 94:
    grades = 1.3
elif percent_gwa >= 92:
    grades = 1.4
elif percent_gwa >= 90:
    grades = 1.5
elif percent_gwa >= 88:
    grades = 1.6
elif percent_gwa >= 86:
    grades = 1.75
elif percent_gwa >= 84:
    grades = 1.8
elif percent_gwa >= 82:
    grades = 1.9
elif percent_gwa >= 80:
    grades = 2.0
elif percent_gwa >= 78:
    grades = 2.1
elif percent_gwa >= 76:
    grades = 2.25
elif percent_gwa >= 74:
    grades = 2.3
elif percent_gwa >= 72:
    grades = 2.4
elif percent_gwa >= 70:
    grades = 2.5
elif percent_gwa >= 68:
    grades = 2.6
elif percent_gwa >= 66:
    grades = 2.75
elif percent_gwa >= 64:
    grades = 2.8
elif percent_gwa >= 62:
    grades = 2.9
elif percent_gwa >= 60:
    grades = 3.0

# Output
print('')
print('=== 1ST SEMESTER RESULTS ===')
print('Percentage GWA:', round(percent_gwa, 2))
print('Status:', 'PASSED' if grades >= 3.0 else 'FAILED')
```

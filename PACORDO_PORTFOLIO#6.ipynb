```python
###### Program Title
print("===================================")
print("         STUDY BUDDY BOT")
print("===================================\n")

# -------------------------
# User Input
# -------------------------
name = input("Enter your name: ").strip()
buddy_name = input("Give your Study Buddy a name: ").strip()

# -------------------------
# List Initialization
# -------------------------
reminders = []

# -------------------------
# Shutdown Phrase
# -------------------------
shutdown_phrase = "I am done, Thank you for your help!"

# -------------------------
# Functions
# -------------------------

# Greeting Function
def show_greeting(user_name, bot_name):
    print("\nHello,", user_name + "!")
    print("I am", bot_name + ", your Study Buddy.")
    print("I can help you with reminders, study tips, and simple suggestions.")
    print('If you want to end our session, say:')
    print('"' + shutdown_phrase + '"\n')

# Menu Function
def show_menu():
    print("Choose an option:")
    print("1 - Add a reminder")
    print("2 - Show reminders")
    print("3 - Remove a reminder")
    print("4 - Get a study tip")
    print("5 - Practice conversation")
    print("6 - End session")

# Add Reminder Function
def add_reminder():
    item = input("Enter a reminder to add: ").strip()
    reminders.append(item)
    print("Reminder added!")

# Show Reminders Function
def show_reminders():
    if len(reminders) == 0:
        print("No reminders yet.")
    else:
        print("\nYour reminders:")
        for i in range(len(reminders)):
            print(str(i + 1) + ".", reminders[i])

# Remove Reminder Function
def remove_reminder():
    if len(reminders) == 0:
        print("There are no reminders to remove.")
        return

    show_reminders()

    # User Input
    choice = input("Enter the number of the reminder to remove: ").strip()

    # Input Validation
    if choice.isdigit():
        index = int(choice) - 1

        # Conditional Statement
        if index >= 0 and index < len(reminders):
            removed = reminders.pop(index)
            print("Removed:", removed)
        else:
            print("Invalid reminder number.")
    else:
        print("Please enter a valid number.")

# Study Tip Function
def study_tip():
    tips = [ "Study in short sessions and take small breaks.", "Review your notes before sleeping.",
 "Practice one topic at a time.", "Drink water and rest properly.", "Use active recall instead of just rereading." ]

    # Display Output
    print("\nStudy Tip:", tips[0])

# Conversation Practice Function
def practice_conversation():
    print("\nLet's practice conversation.")

    # User Input
    question = input("Ask me something simple: ").strip().lower()

    # Conditional Statements
    if "hello" in question:
        print("Hello! Nice to talk with you.")

    elif "how are you" in question:
        print("I am doing well. Thank you for asking.")

    elif "what is your name" in question:
        print("My name is", buddy_name + ".")

    elif "study" in question:
        print("Study one topic at a time and stay consistent.")

    elif shutdown_phrase.lower() in question.lower():
        print("\nPowering down...")
        print("Thank you for studying with me,", name + "!")
        return True

    else:
        print("That is a good question.")
        print("I am still learning, but I support your study journey!")

    return False

# Greeting Display
show_greeting(name, buddy_name)

# Main Loop
while True:

    # Menu Display
    show_menu()

    # User Choice Input
    choice = input("\nEnter your choice: ").strip()

    # Conditional Statements
    if choice == "1":
        add_reminder()

    elif choice == "2":
        show_reminders()

    elif choice == "3":
        remove_reminder()

    elif choice == "4":
        study_tip()

    elif choice == "5":

        # Function Call with Shutdown Condition
        if practice_conversation():
            break

    elif choice == "6":

        # Shutdown Input
        final_phrase = input(
            'Type "' + shutdown_phrase + '" to end the session: '
        ).strip()

        # Conditional Statement
        if final_phrase == shutdown_phrase:
            print("\nPowering down...")
            print("Thank you for using", buddy_name + ",", name + "!")
            break
        else:
            print("Shutdown phrase incorrect.")

    else:
        print("Invalid choice. Please enter a number from 1 to 6.")

    # Display Separator
    print("\n-----------------------------------")
```

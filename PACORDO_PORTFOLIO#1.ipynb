This portfolio presents a calorie counter program that applies basic Python concepts, including:

• Variables

• Expressions

• Type conversion

The program calculates the total calorie intake from selected meals using basic arithmetic operations, demonstrating the practical application of programming in everyday computations.

```python
# User Information
name = input("Enter your name: ")
food = input("Enter a food item: ")
servings = float(input("How many servings did you eat? "))
calories_per_serving = float(input("Calories per serving: "))

# Calculations
total_calories = servings * calories_per_serving

# Output
print("Hello,", name)
print("Food item:", food)
print("Total calories:", total_calories)
```

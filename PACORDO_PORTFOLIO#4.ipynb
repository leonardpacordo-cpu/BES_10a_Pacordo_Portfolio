# Topic Description

This portfolio output presents a mini gacha-style slot machine game inspired by Genshin Impact that simulates a wishing system using random symbols. The user enters a name, starts with a fixed number of “Fate” credits, and chooses whether to continue making wishes. Each pull consumes one credit, and the game continues until the user stops or runs out of credits. A win occurs when all three symbols match, granting bonus credits as a reward. This program uses fundamental programming concepts, including:

• Iteration

• While Loop

• Break Statement

• Conditional Statements

• Functions 

• Variables

• Expressions

• User Input

• Random Values

Overall, this program mainly demonstrates the concept of iteration through the use of a while loop that controls the continuous execution of the game. The loop allows repeated gameplay, where each iteration represents a single “wish” or spin until a stopping condition is met, either when the user decides to stop or when credits are depleted. This shows how loops are essential in creating interactive programs that require repetition, control flow, and user-driven continuation.

```python
# Input Module
import random

# Program Title
print("Spin of Fate: Gacha Wish Simulator")

# User Input
player = input("Enter your name, Traveler: ")

# Variable Initialization
credits = 10
symbols = ["Primogem", "Fate", "Mora", "Vision", "Crown"]

# Function: Generate Random Pull
def spin_machine():
    return [ random.choice(symbols), random.choice(symbols), random.choice(symbols) ]

# Function: Check if Player Gets a Rare Pull
def check_win(result):
    return result[0] == result[1] == result[2]

# Initial Game Introduction
print("\nWelcome, Traveler,", player)
print("You have been granted 10 Intertwined Fate to begin your journey.")
print("Each pull costs 1 Fate.")
print("Match 3 symbols to obtain bonus rewards from the stars.\n")

# Main Game Loop
while credits > 0:

    # User Input: Continue or Stop Pulling
    play = input("Press Enter to wish or type 'stop' to stop: ")

    if play == "stop":
        break

    # Computation: Spend one Fate
    credits -= 1

    # Generate Pull Result
    result = spin_machine()

    
    print("Your wish revealed:", result[0], "|", result[1], "|", result[2])

    if check_win(result):
        credits += 5
        print("The stars align! You obtained a rare blessing! +5 Fate")
    else:
        print("No resonance... the universe did not respond this time.")

    print("Remaining Fate:", credits)
    print()

if credits == 0:
    print("You have run out of Fate... There are not enough Primogems to make another wish.")
    print("The gacha banner fades into the distance, unanswered.")
else:
    print("Thank you for playing, Traveler,", player)
    print("May your next wishes be blessed by the stars.")
```

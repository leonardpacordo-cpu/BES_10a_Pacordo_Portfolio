This portfolio output presents a personal branding generator that creates customized usernames, initials, stylish name displays, and formatted nicknames using different string manipulation techniques in Python. The user enters their full name, nickname, and favorite word, which are then processed through various string methods to clean spaces, convert text formats, and generate personalized branding details. The program also performs simple text analysis, such as counting specific characters and measuring string lengths. This program uses fundamental programming concepts, including:

• Variables

• User Input

Overall, this program mainly demonstrates the concept of string manipulation through the use of different string methods and operations that process and format user input into personalized outputs. It shows how strings can be cleaned, modified, combined, and analyzed to create organized and visually appealing text-based information. This highlights the practical application of Python in handling textual data for customization, formatting, and personal branding purposes.

```python
# Display Program Title
print("Personal Branding Generator\n")

# User Input
full_name = input("Enter your full name: ")
nickname = input("Enter your nickname: ")
favorite_word = input("Enter your favorite word: ")

# Space Cleaner
full_name_clean = full_name.strip()
nickname_clean = nickname.strip()
favorite_word_clean = favorite_word.strip()

# String Formatting
name_title = full_name_clean.title()
nickname_lower = nickname_clean.lower()
favorite_upper = favorite_word_clean.upper()

# Splitting Name into Parts
name_parts = full_name_clean.split()

# Extracting First and Last Name
first_name = name_parts[0]
last_name = name_parts[-1]

# Initial Creator
first_initial = first_name[0].upper()
last_initial = last_name[0].upper()
initials = first_initial + last_initial

# Username Generator
username = (first_name.lower() + "_" + nickname_lower + str(len(full_name_clean)))

# Stylish Display Formatter
stylish_name = "*" * 2 + " " + name_title + " " + "*" * 2

# Nickname Formatter
nickname_style = "~" + nickname_clean.title() + "~"

# Uppercase and Lowercase Converter
name_upper = full_name_clean.upper()
name_lower = full_name_clean.lower()

# String Length Computation
name_length = len(full_name_clean)
nickname_length = len(nickname_clean)

# Character Counting
a_count = favorite_word_clean.lower().count("a")
e_count = favorite_word_clean.lower().count("e")

# Display Personal Branding Information
print("\n--- Personal Branding Profile ---")
print("Stylish Name:", stylish_name)
print("Generated Username:", username)
print("Initials:", initials)
print("Formatted Nickname:", nickname_style)

# Display String Formatting Results
print("\n--- Text Formatting ---")
print("Uppercase Name:", name_upper)
print("Lowercase Name:", name_lower)
print("Favorite Word in Uppercase:", favorite_upper)

# Display String Details
print("\n--- String Details ---")
print("Name Length:", name_length)
print("Nickname Length:", nickname_length)

# Final Summary
print("\n--- Summary ---")
print(name_title + " created the username '" + username + "' and uses the nickname " + nickname_style + ".")
```
